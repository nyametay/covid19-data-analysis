### Q.3) In which Region, maximum number of Confirmed cases were recorded ?

```python
df.groupby('Region')['Confirmed'].sum().reset_index().sort_values('Confirmed', ascending=False)
```

### Q.4) In which Region, minimum number of Deaths cases were recorded ?

```python
filtered = df.groupby('Region')['Deaths'].sum().reset_index().sort_values('Deaths', ascending=True)
filtered
```

```python
filtered[filtered.Deaths == 0]
```
